##### **Task 2**

```python
def rename_my_file(old_file_name):
    global change
    import os, glob

    try:
        if old_file_name[-3:].lower()=='txt':
            new_file_name=change[old_file_name[:3]]+old_file_name[3:]
            dir='os-collaboration-exercise-data'
            old_file_path=os.path.abspath(os.path.join('..',dir,old_file_name))
            new_file_path=os.path.abspath(os.path.join('..',dir,new_file_name))
            os.rename(old_file_path,new_file_path)
            return f'Changed: {old_file_name} ----> {new_file_name}'
        
        elif old_file_name[0].isdigit():
            return f'Seems like file is already changed!'
            
        else:
            return 'Oops! Only .txt files please!'

    except:
        return 'Error! The file may have already been changed before or doesn\'t exist!'
    
print(rename_my_file('apr-01_melbourne.txt'))
```

##### **Task 7**

```python
import glob, os, shutil

cities=['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

for i in cities:
    req_dir=os.path.abspath(os.path.join('..','os-collaboration-exercise-data',f'{i}'))
    city_files=glob.glob(os.path.abspath(os.path.join('..','os-collaboration-exercise-data',f'*{i.lower()}.txt')))
    try:
        for j in city_files:
            shutil.move(j,req_dir)
            print(f'Moved {j} -----> {req_dir}')
    except: #file already exists
        continue

else:
    print('Done!') #to print after for loop runs throughout
```
